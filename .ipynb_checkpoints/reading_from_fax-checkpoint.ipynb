```python
# import module
from pdf2image import convert_from_path
import os
from tqdm import tqdm
```

```python
def convert_pdf_image(pdf_path, save_folder, filename):
    
    # Store Pdf with convert_from_path function
    images = convert_from_path(pdf_path, poppler_path = r'C:\Nisarg Projects\Fax_Reading\poppler-23.07.0\Library\bin')
    
    if os.path.isdir(os.path.join(save_path, filename[:-4])):
        pass
    else:
        os.mkdir(os.path.join(save_path, filename[:-4]))
    
    for i in range(len(images)):

        # Save pages as images in the pdf
        images[i].save(os.path.join(os.path.join(save_path, filename[:-4]), 'page_'+ str(i) +'.png'), 'PNG')
```

```python
pdf_folder = r'C:\Nisarg Projects\Fax_Reading\PDFs'
save_folder = r'C:\Nisarg Projects\Fax_Reading\Images'
```

```python
for filename in tqdm(os.listdir(pdf_folder)):
    pdf_path = os.path.join(pdf_folder, filename)
    convert_pdf_image(pdf_path, save_folder, filename)
```

```python
import cv2
import matplotlib.pyplot as plt

def draw_bounding_boxes(image_path, box_width, box_height):
    # Load the image
    image = cv2.imread(image_path)

    # Check if the image is loaded successfully
    if image is None:
        print("Error: Unable to load the image.")
        return

    # Get the height and width of the image
    height, width, _ = image.shape

    # Define the color of the bounding boxes (BGR format)
    box_color = (0, 255, 0)  # Green color

    # Define the starting coordinates (x, y)
    x_start, y_start = 0, 400

    # Loop through the image and draw bounding boxes at regular intervals
    x, y = x_start, y_start
    while y + box_height <= height:
        cv2.rectangle(image, (x, y), (x + box_width, y + box_height), box_color, 2)
        x += box_width
        if x + box_width > width:
            x = x_start
            y += box_height

    # Display the image with bounding boxes
    plt.imshow(image)
#     cv2.imshow('Image with Bounding Boxes', image)
#     cv2.waitKey(0)
#     cv2.destroyAllWindows()


image_path = r'C:\Nisarg Projects\Fax_Reading\Images\pdf_1\page_1.png'

draw_bounding_boxes(image_path, box_width, box_height)
```

```python
def find_coordinates_inside_bbox(result_dict, bounding_boxes):
    result = {i: [] for i in range(len(bounding_boxes))}

    for bbox_index, bbox in enumerate(bounding_boxes):
        x, y, width, height = bbox
        
        coordinates = list(result_dict.values())
        
        for coord_set in coordinates:
            for coord in coord_set:
                px, py = coord

                if x <= px < x + width and y <= py < y + height:
                    result[bbox_index].append(list(result_dict.keys())[list(result_dict.values()).index(coord_set)])
                    print(coord_set)
                    #print(list(result_dict.keys())[list(result_dict.values()).index(coord_set)])
                    break

    return result
```
